Data Preparation

```python
cars_df["companyname"]=cars_df["CarName"].str.split(' ').str[0]


```

```python
#dropping columns which are not relevant.
cars_df.drop(['CarName'],axis=1,inplace=True)
cars_df.drop(['car_ID'],axis=1,inplace=True)
cars_df.head(20)
```

```python
#listing the unique set of columns values
cars_df.companyname.unique()

```

We can see above that there are few mispelled company names like maxda(mazda),porcdhce(porsche),toyouta(toyota),vw(volkswagen),vokswagen(volkswagen).

```python
cars_df.companyname.replace('maxda','mazda', inplace = True)
cars_df.companyname.replace('porcshce','porsche', inplace = True)
cars_df.companyname.replace('vokswagen','volkswagen', inplace = True)
cars_df.companyname.replace('toyouta','toyota', inplace = True)
cars_df.companyname.replace('vw','volkswagen', inplace = True)
cars_df.companyname.unique()
```

```python
#making all the company names in lower case (specifically for 'Nissan')

cars_df.companyname = cars_df.companyname.str.lower()
cars_df.companyname.unique()
```

```python
round(cars_df.isnull().sum()/len(cars_df.index), 2)*100 # NO missing values in the data set.
cars_df =cars_df.drop_duplicates() # dropping duplicate values..
```


Splitting Train and Test data

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(cars_df, train_size = 0.7, test_size = 0.3, random_state = 100)
cars_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
#scaling few numerical values in the data this is to standardise the value across the data set.
scaler = MinMaxScaler()
numerical_var = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'carheight', 'horsepower','stroke','carlength','carwidth','compressionratio','peakrpm','citympg','highwaympg','price']
df_train[numerical_var] = scaler.fit_transform(df_train[numerical_var])
df_train.describe()
```


Modelling (using RFE)

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
y_train = df_train.pop('price')
X_train = df_train
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(estimator=lm, n_features_to_select=10)
# Fit RFE to the training data
rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_] #columns support or to be considered under RFE model
col
```

Checking for multi-collinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Predictions

```python
#Scaling the uneven data values into a standardised format using max/mix scaling
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'carheight', 'horsepower','stroke','carlength','carwidth','compressionratio','peakrpm','citympg','highwaympg','price']

df_test[num_vars] = scaler.fit_transform(df_test[num_vars])

#df_test.head()
y_test = df_test.pop('price')
X_test = df_test

X_test.describe()


X_train_new.columns
X_train_new = X_train_new.drop('const',axis=1)
# Creating X_test_new dataframe by dropping variables from X_test


# Adding a constant variable
#X_test_new = sm.add_constant(X_test_new)
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable
#X_test_new = sm.add_constant(X_test_new)
X_test_new = sm.add_constant(X_test_new)


#predicting 'price' on the test data set
y_pred = lm.predict(X_test_new)

# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)


from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(y_test, y_pred))
print('RMSE:',rmse)

from sklearn.metrics import r2_score
r2=r2_score(y_test, y_pred)
print('r2_score:',r2)


```



    R- Squared: This model explains 87% of the variance
    F-stats and Prob(F-stats): 239.0 & 8.12e-66. Both values show that this model is significant.
    p-values of all the dependent variables is less than its significance value(0.05)
